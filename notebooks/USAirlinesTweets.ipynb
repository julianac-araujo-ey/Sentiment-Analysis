<center><h1><div class="top_section">1. O que é Análise de Sentimentos?</div></h1></center>

Dados não estruturados como postagens nas mídias sociais requerem processamento para gerar insights. A análise de sentimentos (mineração de opinião ou IA de emoções) é uma tarefa de Processamento de Linguagem Natural (PNL), que alia linguística computacional e mineração de texto para identificar e extrair sentimentos e emoções. Na maioria dos casos foca-se na polaridade (positivo, negativo, neutro), mas também pode-se extender o problema para detecção de sentimentos e emoções específicas (felicidade, frustração, etc), urgência (urgente, não urgente) e até intenções (interessado vs. não interessado).

A Análise de Sentimentos transforma informações não-estruturadas em dados estruturados de opiniões sobre produtos, serviços, marcas, política ou qualquer tópico sobre o qual as pessoas expressem suas impressões. Estes dados são úteis em análises de marketing, relações públicas, análises e feedback de produtos e atendimento ao cliente. Como a tarefa de classificação de sentimento é, na maioria das vezes um processo automatizado, ganha-se escalabilidade e agilidade na extração de informações.


<center><h1><div class="top_section">2. Onde é aplicada?</div></h1></center>

Existem diversas formas de aplicação:

- Análise de texto, livros, mensagens, postagens, poemas, notícias, mídias sociais...
- Ferramentas de suporte à escrita (correção automática, preenchimento automático, verificação ortográfica)
- Tomada de decisão de investimento financeiro (através da análise do sentimento do mercado)
- Formulação de políticas (análise da reação pública em assuntos específicos). A administração Obama usou a análise de sentimentos para prever a resposta pública aos seus anúncios.
- Aplicativos de voz
- Mecanismos de busca
- Business Development e Business Inteligence
- Videogames e Realidade Virtual
- Assistentes de voz (Cortana, Siri, Alexa, Athena, Mycroft, Aido, Cubic, Hound, Databot, AIVC, Butleroy, Robin, Lyra, Maluuba, Dragon, Lucida, Bixby, Vlingo, Silvia para citar alguns)



<center><h1><div class="top_section">3. Abordagens para o problema</div></h1></center>

Os algorítmos comumente utilizados na determinação computacional de sentimentos textuais podem ser agrupados em três categorias:

* <strong> Algorítmos baseados em léxicos de sentimento ou não supervisionados</strong>: envolvem o cálculo da polaridade de um texto a partir de orientação semântica das palavras contidas por meio de listas e dicionários de palavras associadas a sentimentos específicos. A eficiência dos métodos léxicos está diretamente relacionada a generalização do vocabulário utilizado. 
* <strong> Algoritmos automáticos ou supervisionados</strong>: utilizam o princípio da inferência denominado indução, para tirar conclusões a partir de um conjunto de exemplos (machine learning, deep learning). Nestes casos a acurácia do método de análise será proporcional a quantidade e qualidade dos dados de treinamento. 
* <strong> Algoritmos Hibridos </strong>: combinam abordagens supervisionadas e não supervisionadas.


<center><h1><div class="top_section">4. Existe 100% de acurácia em Análise de Sentimento?</div></h1></center>

A realização de análise de sentimentos possui diversos desafios, incluindo:

- Gírias, falhas na digitação e “internetês” (abreviações e emojis);
- Sarcasmos, metáforas e ironia;
- Ambiguidade e polissemia.
- Multilinguagem

**Nota:** Uma forma simples e eficiente para endereçar a análise de sentimentos em diferentes
idiomas é combinar o poder dos métodos já existentes em inglês com a eficiência dos
tradutores automáticos. Embora, gírias e expressões linguísticas sejam complexas de se traduzir, o resultado tende a ser satisfatório em muitas aplicações.

<strong><center><h1><div class="top_section">6. Estudo de Caso: VADER</div></h1></center></strong>

VADER (Valence Aware Dictionary and sEntiment Reasoner) é uma ferramenta de análise de sentimento baseada em regras e léxico que é especificamente sintonizada com os sentimentos expressos nas redes sociais onde se tem um contexto vago, textos curtos, gírias e emoticons. 

**Nota:** VADER, que será utilizado aqui, encontra-se disponível no framework NLTK (Natural Language Toolkit).

**PROS:**

- Reduz o esforço manual de rotulação.
- Funciona bem com texto do tipo de mídia social, mas generaliza facilmente para vários domínios, dependendo apenas dos requisitos de negócios e do conhecimento de domínio
- A inferência é rápida o suficiente para usar online com dados de streaming. Devido a rotulação semântica e as regras, VADER é computacionalmente econômico especialmente comparado com algoritmos de aprendizado de máquina e aprendizado profundo que requerem operação maciça para incorporação e treinamento

**CONTRAS:**

- A precisão da análise às vezes pode não ser grande
- VADER ignora o contexto em que grande maioria das palavras são usadas 

Para calcular a pontuação sentimental do texto, VADER escaneia-o em busca de características sentimentais conhecidas, modifica a intensidade e a polaridade de acordo com regras ("very happy" é mais positivo que apenas "happy"), resume as pontuações de características encontradas dentro do texto e normaliza o score final em (-1, 1) usando a função:

$$
score  = \frac{x}{\sqrt{x^2 + \alpha}}, 
$$

Onde x = soma da pontuação (valência) das palavras constituintes do texto, e  $\alpha$ = Constante de normalização (default = 15).

Note que, à medida que x cresce, fica cada vez mais próximo de -1 ou 1. Para efeito semelhante, se houver muitas palavras no documento ao qual você está aplicando VADER, você obtém uma pontuação próxima de -1 ou 1. Desse modo, fica fácil visualizar porque o algorítmo funciona melhor em documentos curtos, como tweets.

Além do score do texto, VADER também retorna a porcentagem de características de sentimento positivo, negativo e neutro. O dicionário léxico do VADER contém cerca de 7.500 características de sentimento no total e qualquer palavra não listada no dicionário será pontuada como "0: Neutro".

Em VADER, os desenvolvedores incorporaram várias regras heurísticas que tratam dos casos de pontuação, capitalização, advérbios e conjunções contrastivas. Além do léxico de sentimento, existem estruturas que são neutras inerentemente, mas podem mudar a polaridade do sentimento (como "not" e "but") ou modificar a intensidade de toda a sentença (como "very" e "extremely"). 

**NOTA para leitores avançados**: Essas heurísticas vão além do que normalmente seria capturado em um modelo bag-of-words. Elas incorporam relações sensíveis à ordem das palavras entre os termos.

Todas as características léxicas foram previamente classificadas por polaridade e intensidade em uma escala de "-4: Extremamente Negativo" a "+4 Extremamente Positivo" por 10 avaliadores humanos. Mais detalhes [aqui](http://comp.social.gatech.edu/papers/icwsm14.vader.hutto.pdf).

Percebemos que algumas sentenças podem ser classificadas como o julgamento humana.

Mas para frases ambíguas (ou sarcásticas) VADER pode não funcionar bem. Exemplos:

- Wanted to get my bag benefit, but instead get $25 pricing on all three tickets. When adding a card, MP Visa is only option. @united

- @SouthwestAir I'm an A-list Preferred customer and very loyal SWA. However, your inflight wifi is no better today than it was 4 years ago.

O problema é que VADER usa termos simples para expressar sentimentos. No entanto, fatores culturais, nuances e contextos linguísticos diferentes tornam difícil classificar uma sequência de texto escrito em positiva ou negativa. Ainda, o fato de que pessoas muitas vezes discordam sobre o sentimento do texto ilustra o quão grande é essa tarefa para os algorítmos. 

Abordagens que analisam o sentimento com base em como as palavras compõem o significado de frases mostraram melhores resultados, mas incorrem em uma sobrecarga de anotação adicional.
